```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
file_path = '/content/drive/MyDrive/dataset/Top Indian Places to Visit.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

EDA


 Distribution of tourist spots across different zones, states, and cities

```python
plt.figure(figsize=(20, 8))
sns.countplot(x='Type', data=df, order=df['Type'].value_counts().index, palette='viridis')
plt.title('Distribution of Tourist Spots Based on Types')
plt.xlabel('Type')
plt.ylabel('Number of Tourist Spots')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(x='Significance', data=df, order=df['Significance'].value_counts().index, palette='viridis')
plt.title('Distribution of Tourist Spots Based on Significance')
plt.xlabel('Significance')
plt.ylabel('Number of Tourist Spots')
plt.xticks(rotation=45)
plt.show()
```

Relationship between establishment year and Google review ratings

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(x='Establishment Year', y='Google review rating', data=df, palette='viridis')
plt.title('Relationship Between Establishment Year and Google Review Ratings')
plt.xlabel('Establishment Year')
plt.ylabel('Google Review Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Entrance Fee in INR', y='Google review rating', data=df, palette='viridis')
plt.title('Correlation Between Entrance Fee and Google Review Ratings')
plt.xlabel('Entrance Fee in INR')
plt.ylabel('Google Review Rating')
plt.show()
```
