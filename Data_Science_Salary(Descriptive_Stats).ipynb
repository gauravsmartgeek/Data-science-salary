```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "/content/drive/MyDrive/dataset/data_science_salaries.csv"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
unique_values = {
    col: data[col].unique() for col in ['job_title', 'experience_level', 'employment_type', 'work_models', 'work_year', 'employee_residence', 'salary_currency', 'company_location', 'company_size']
}
unique_values
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['salary_in_usd'], kde=True)
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=data)
plt.title('Salary vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=data)
plt.title('Salary vs Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='work_models', y='salary_in_usd', data=data)
plt.title('Salary vs Work Models')
plt.xlabel('Work Models')
plt.ylabel('Salary in USD')
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
